```python
%cd ../TreePPL/treeppl-python/examples
import sys

sys.path.append("..")

import treeppl
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

total_samples = 500
sofar_samples = 0

tree = treeppl.Tree.load("trees/bing_small.phyjson", format="phyjson")
from timeit import default_timer as timer

samples = None
logz = []
times = []

with treeppl.Model(filename="crbd.tppl", samples=100, resample="likelihood") as crbd:
    while sofar_samples < total_samples:
        start = timer()
        res = crbd(tree=tree)
        samples = pd.concat([samples, pd.DataFrame({
            'lambda': res.subsample(10),
            'lweight': res.norm_const,
        })])
        end = timer()
        sofar_samples+=10
        logz.append(res.norm_const)
        times.append(end - start)
        print(".", end="")
    
print(logz)
print(times)     
        
samples = None
sofar_samples = 0
logz_aligned = []
times_aligned = []

with treeppl.Model(filename="crbd-aligned.tppl", samples=100, resample="likelihood") as crbd:
    while sofar_samples < total_samples:
        start = timer()
        res = crbd(tree=tree)
        samples = pd.concat([samples, pd.DataFrame({
            'lambda': res.subsample(10),
            'lweight': res.norm_const,
        })])
        end = timer()
        sofar_samples+=10
        logz_aligned.append(res.norm_const)
        times_aligned.append(end - start)
        print(".", end="")
    

print(logz_aligned)
print(times_aligned)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np



# Graph A: Violin Plot
plt.figure(figsize=(10, 6))
sns.violinplot(data=[logz, logz_aligned])
plt.xticks([0, 1], ['logz', 'logz_aligned'])
plt.title('Distribution of logz and logz_aligned')
plt.show()

# Graph B: Bar Chart with Error Bars
mean_times = np.mean(times)
mean_times_aligned = np.mean(times_aligned)
error_times = np.std(times)
error_times_aligned = np.std(times_aligned)

plt.figure(figsize=(10, 6))
plt.bar(['times', 'times_aligned'], [mean_times, mean_times_aligned], yerr=[error_times, error_times_aligned], capsize=10)
plt.title('Comparison of times and times_aligned')
plt.show()

import matplotlib.pyplot as plt
import numpy as np  # For standard deviation

# Assuming 'times' and 'times_aligned' are your data lists


import matplotlib.pyplot as plt
import numpy as np
import matplotlib




plt.figure()
bars = plt.barh(['Aligned program', 'Unaligned program'], [sum(times_aligned), sum(times)], 
                xerr=[np.std(times_aligned), np.std(times)], capsize=10, 
                edgecolor=['blue', 'green'], linewidth=1.5, fill=False)  # Hollow bars
plt.title('Comparison of running times in seconds (lower is better)')

# Save the plot as an SVG file
plt.savefig('horizontal_bar_chart.svg', format='svg')

plt.show()



```
